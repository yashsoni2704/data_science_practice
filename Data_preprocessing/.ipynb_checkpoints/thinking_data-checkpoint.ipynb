```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
data = load_data("store_data.json")
```

```python
def clean_data(data):
    #cleaning of data
    text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}
    unique_user = set()
    cleaned_data = []
    for user in data:
        raw_rating = user["rating"]
        if isinstance(raw_rating,str):
            raw_rating = raw_rating.strip().lower()
            if raw_rating in text_to_num:
                raw_rating = text_to_num[raw_rating]
            try:
                raw_rating = float(raw_rating)
            except:
                pass
        if isinstance(raw_rating,float):
            raw_rating = int(raw_rating)
        user["rating"] = raw_rating
        # print(user["rating"])

        # handling missing data
        raw_age = user.get("age")
        if (raw_age == None):
            user["age"] = None
        # print(user)


        #deduplication
        if (user["name"].strip().lower() in unique_user):
            continue
        unique_user.add(user["name"].strip().lower())
        cleaned_data.append(user)
    return cleaned_data
```

```python
cleaned_data = clean_data(data)
```

```python
def get_insights(data):
    #getting average rating
    total_rating = 0
    for user in data:
        total_rating = total_rating + user["rating"]
    print(total_rating)
    print(len(data))
    print(f"avg rating = {total_rating/(len(data))}")

    #getting how much percentage of users gave poor ratings
    poor_rating = 0
    for user in data:
        if user["rating"] < 3:
            poor_rating += 1
    print(poor_rating)
    print(len(data))
    print(f"% of poor rating = {poor_rating/len(data) * 100}%")
```

```python
get_insights(cleaned_data)
```

```python
#build recommendation feature
def get_recommendation(cleaned_data):
    recommendation = []
    for user in data:
        current_recommend = {}
        current_recommend["name"] = user["name"]
        if(user["rating"] >= 4):
            current_recommend["brand"] = "Apple"
        else:
            current_recommend["brand"] = "Samsung"
        recommendation.append(current_recommend)
    return recommendation
```
